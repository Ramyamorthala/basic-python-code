```python
from matplotlib import pyplot as plt # for visualization
import seaborn as sns #for advanced vis /statistical visualization

%matplotlib inline 

import warnings
warnings.filterwarnings('ignore')
```

```python
vis = sns.jointplot(data = movies, x = 'Critic rating' , y = 'Audience rating')
plt.show()
# Audience rating is more dominant then critics rating
# Based on this we find out as most people are most liklihood to watch audience rating & less likely to wathc critics rating
# let me explain the excel - if you filter audience rating & critic rating. critic rating has very low values compare to audience rating 
```

```python
vis = sns.jointplot(data = movies, x = 'Critic rating' , y = 'Audience rating',kind = 'scatter')
plt.show()
```

```python
vis = sns.jointplot(data = movies, x = 'Critic rating' , y = 'Audience rating',kind = 'reg')
plt.show()
```

```python
vis = sns.jointplot(data = movies, x = 'Critic rating' , y = 'Audience rating',kind = 'hex') # kind helps in plotting the graph differnt types
plt.show()
```

```python
vis = sns.jointplot(data = movies, x = 'Critic rating' , y = 'Audience rating',kind = 'kde' ,cmap = 'Greens') #kde= kernel density plots
plt.show()
```

```python
plt.hist(movies.Budgetmillions) #univaritae analysis
plt.show()
```

```python
graph = sns.lmplot(data = movies , x ='Critic rating', y = 'Audience rating' ,
                   hue = 'Genre' , fit_reg = False)
plt.show() #hue helps in identifing the speicific catergory
```

```python
graph = sns.lmplot(data = movies , x = 'Critic rating', y = 'Audience rating' , fit_reg = True , hue = 'Genre')
plt.show()
```

```python
b = sns.boxplot(data = movies , x = 'Genre' , y = 'Critic rating', hue= 'Genre')
plt.show()
```

```python
v = sns.violinplot(data = movies , x = 'Genre' , y = 'Critic rating',hue = 'Genre')
plt.show()
```

```python
D = sns.boxplot(data = movies[movies.Genre == 'Drama'] , x = 'Year' , y = 'Audience rating')
plt.show()
```

```python
#Facetgrid (#multivariate analysis)

g = sns.FacetGrid(movies , row = 'Genre' , col = 'Year' , hue = 'Genre')
plt.show()
```

```python
#so to fill the facetgrids , we need to use map function and hence maps are plotted in the graph


g = sns.FacetGrid(movies , row = 'Genre' , col = 'Year' , hue = 'Genre')
g = g.map(plt.scatter , 'Critic rating' ,'Audience rating')
plt.show()                               #with map function, scatterplots are mapped in facetgrids
```

```python
F = sns.FacetGrid(movies , row = 'Genre' , col = 'Year' , hue = 'Genre')
F = F.map(plt.hist , 'Budgetmillions') #histograms are mapped in facetgrids
plt.show()
                  
```
